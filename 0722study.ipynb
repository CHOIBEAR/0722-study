```python
# 1. 인사 프로그램
# 설명: 사용자로부터 이름(name)과 나이(age)를 입력받아
# 안녕하세요, {name}님! 내년엔 {age+1}살이시군요!
# 와 같은 형식으로 출력하세요.
# 입력 예시:
# 홍길동
# 25
# 출력 예시:
# 안녕하세요, 홍길동님! 내년엔 26살이시군요!

Name = input("이름을 입력하세요: ")  # 사용자로부터 이름 입력
age = int(input("나이를 입력하세요: "))  # 사용자로부터 나이 입력

print(f"안녕하세요, {Name}님! 내년엔 {age + 1}살이시군요!")  # 인사 메시지 출력
```

```python
# 2. 사칙연산 계산기
# 설명: 두 정수 a, b를 입력받아 덧셈·뺄셈·곱셈·나눗셈(몫과 나머지)을 차례대로 출력하세요.
# 입력 예시:
# 7 3
# 출력 예시:
# 7 + 3 = 10
# 7 - 3 = 4
# 7 * 3 = 21
# 7 // 3 = 2, 나머지 = 1

a=int(input("첫 번째 정수를 입력하세요: "))  # 사용자로부터 첫 번째 정수 입력
b=int(input("두 번째 정수를 입력하세요: "))  # 사용자로부터 두 번째 정수 입력

print(f"{a} + {b} = {a + b}")  # 덧셈 결과 출력
print(f"{a} - {b} = {a - b}")  # 뺄셈 결과 출력
print(f"{a} * {b} = {a * b}")  # 곱셈 결과 출력
print(f"{a} // {b} = {a // b}, 나머지 = {a % b}")  # 나눗셈(몫과 나머지) 결과 출력   
```

```python
# 3. 학점 계산기 (조건문 연습)
# 설명: 0~100 점 사이의 점수(score)를 입력받아
# 90점 이상 → A
# 80점 이상 → B
# 70점 이상 → C
# 60점 이상 → D
# 그 외 → F
# 를 출력하세요.
# 입력 예시:
# 85
# 출력 예시:
# B

score = int(input("점수를 입력하세요 (0~100): "))  # 사용자로부터 점수 입력
if score >= 90:  # 90점 이상    
    print("A")  # A 학점 출력
elif score >= 80:  # 80점 이상 
    print("B")  # B 학점 출력
elif score >= 70:  # 70점 이상 
    print("C")  # C 학점 출력
elif score >= 60:  # 60점 이상
    print("D")  # D 학점 출력
else:  # 그 외
    print("F")
    
```

```python
# 4. 1부터 n까지 합 구하기 (for문)
# 설명: 정수 n을 입력받아 1부터 n까지의 합을 계산하여 출력하세요.
# 입력 예시:
# 10
# 출력 예시:
# 1 + 2 + … + 10 = 55

n = int(input("정수를 입력하세요: "))  # 사용자로부터 정수 n 입력
total = 0  # 합계를 저장할 변수 초기화  
for i in range(1, n + 1):  # 1부터 n까지 반복
    total += i  # 합계에 현재 숫자 추가
print(f"1 + 2 + … + {n} = {total}")  # 결과 출력
```

```python
# 5.사용자로부터 정수를 하나 입력받아,
# 1부터 해당 숫자까지의 합을 구해서 출력하는 프로그램을 작성하세요.
# 단, while 문을 사용하세요.

n = int(input("정수를 입력하세요: "))  # 사용자로부터 정수 n 입력
total = 0  # 합계를 저장할 변수 초기화
i = 1  # 반복 변수 초기화
while i <= n:  # 1부터 n까지 반복
    total += i  # 합계에 현재 숫자 추가
    i += 1  # 반복 변수 증가
print(f"1 + 2 + … + {n} = {total}")  # 결과 출력

```

```python
# 6.사용자로부터 비밀번호를 입력받아,
# 올바른 비밀번호("python123")가 입력될 때까지 반복해서 묻는 프로그램을 작성하세요.
# 입력한 비밀번호가 맞으면 "로그인 성공!"을 출력하고 종료하세요.
# (힌트: while문과 input()을 사용하세요.)

pwd = "python123"  # 올바른 비밀번호 설정
user_input = ""  # 사용자 입력 초기화   
while user_input != pwd:  # 올바른 비밀번호가 입력될 때까지 반복
    user_input = input("비밀번호를 입력하세요: ")  # 사용자로부터 비밀번호 입력
if user_input == pwd:  # 비밀번호가 일치하는지 확인
    print("로그인 성공!")  # 로그인 성공 메시지 출력
```

```python
#7. 문자열 뒤집기
# 설명: 문자열을 입력받아 역순으로 뒤집어 출력하세요.
# 입력 예시:
# Python
# 출력 예시:
# nohtyP

string = input("문자열을 입력하세요: ")  # 사용자로부터 문자열 입력
뒤집힌문자 = string[::-1]  # 문자열을 역순으로 뒤집기
print(뒤집힌문자)  # 뒤집힌 문자열 출력
```

```python
# 8. 학생 점수 평균 (딕셔너리 연습)
# 설명: 학생 이름과 점수를 쉼표(,)로 구분해 여러 줄로 입력받다가, 빈 줄이 들어오면 종료합니다. 입력된 데이터를 딕셔너리로 관리한 뒤, 학생별 점수를 출력하고 전체 학생의 평균 점수를 계산하여 출력하세요.
# 입력 예시:
# Alice,90
# Bob,75
# Charlie,88

# 출력 예시:
# Alice: 90
# Bob: 75
# Charlie: 88
# 전체 평균: 84.33333333333333
학생 = {}  # 학생 정보를 저장할 딕셔너리 초기화
while True:  # 무한 루프 시작
    입력 = input("학생 이름과 점수를 입력하세요 (이름,점수). 빈 줄로 종료: ")  # 사용자로부터 입력 받기
    if 입력 == "":  # 빈 줄 입력 시 종료
        break  # 루프 종료
    이름, 점수 = 입력.split(",")  # 입력을 이름과 점수로 분리
    학생[이름] = int(점수)  # 딕셔너리에 학생 정보 저장
    for 이름, 점수 in 학생.items():  # 딕셔너리의 각 항목을 순회
        print(f"{이름}: {점수}")  # 학생 이름과 점수 출력
    전체_평균 = sum(학생.values()) / len(학생)  # 전체 평균 계산
    print(f"전체 평균: {전체_평균}")  # 전체 평균 출력
```

```python
# 9. 소수 판별기 (함수 작성)
# 설명: 정수 n을 입력받아 소수인지 아닌지 판별하는 함수 소수판별기(n)을 작성하세요. 결과에 따라
# 17는(은) 소수입니다.
# 또는
# 18는(은) 소수가 아닙니다.
# 를 출력하세요.
# 입력 예시:
# 17

def 소수판별기(n):  # 소수 판별 함수 정의
    if n < 2:  # 2보다 작은 수는 소수가 아님
        return False
    for i in range(2, int(n**0.5) + 1):  # 2부터 n의 제곱근까지 반복
        if n % i == 0:  # 나누어 떨어지면 소수가 아님
            return False
    return True  # 소수인 경우 True 반환
n = int(input("정수를 입력하세요: "))  # 사용자로부터 정수 입력
if 소수판별기(n):  # 소수 판별 함수 호출
    print(f"{n}는(은) 소수입니다.")  # 소수인 경우 메시지 출력
else:
    print(f"{n}는(은) 소수가 아닙니다.")
```

```python
# 9.두 개의 정수를 입력받아,
# 두 수의 합을 계산해서 반환하는 함수 add_numbers()를 작성하세요.
# 그리고 그 함수를 호출해서 결과를 출력하세요.

def add_numbers(a, b):  # 두 수의 합을 계산하는 함수 정의
    return a + b  # 두 수의 합 반환
a = int(input("첫 번째 정수를 입력하세요: "))  # 사용자로부터 첫 번째 정수 입력
b = int(input("두 번째 정수를 입력하세요: "))  # 사용자로부터 두 번째 정수 입력
result = add_numbers(a, b)  # 함수 호출하여 결과 저장
print(f"{a} + {b} = {result}")  # 결과 출력
```

```python
# 10. 종합 문제: 과일 장바구니
# 설명: 사용자로부터 과일 이름과 수량을 공백 구분으로 여러 번 입력받습니다. (예: apple 3) 빈 줄이 들어오면 입력 종료 후,
# 1) 각 과일별 총 수량을 출력
# 2) 전체 과일 개수를 합산하여 출력
# 입력 예시:
# 사과 3
# 바나나 2
# 사과 1
# 오렌지 5

# 출력 예시:
# 사과: 4
# 바나나: 2
# 오렌지: 5
# 전체 과일 개수: 11

과일_장바구니 = {}  # 과일 정보를 저장할 딕셔너리 초기화
while True:  # 무한 루프 시작
    입력 = input("과일 이름과 수량을 입력하세요 (이름 수량). 빈 줄로 종료: ")  # 사용자로부터 입력 받기
    if 입력 == "":  # 빈 줄 입력 시 종료
        break  # 루프 종료
    과일, 수량 = 입력.split()  # 입력을 과일과 수량으로 분리
    수량 = int(수량)  # 수량을 정수로 변환  
    if 과일 in 과일_장바구니:  # 과일이 이미 존재하는 경우
        과일_장바구니[과일] += 수량  # 기존 수량에 추가
    else:  # 과일이 새로 입력된 경우
        과일_장바구니[과일] = 수량  # 새로운 과일 추가
for 과일, 수량 in 과일_장바구니.items():  # 딕셔너리의 각 항목을 순회
    print(f"{과일}: {수량}")  # 과일 이름과 수량 출력
전체_과일_개수 = sum(과일_장바구니.values())  # 전체 과일 개수 합산
print(f"전체 과일 개수: {전체_과일_개수}")  # 전체 과일 개수 출력
```
